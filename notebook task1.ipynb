# 📊 Student Data Cleanliness Analysis
In this notebook, we are investigating the `student-mat.csv` dataset to check for any missing values before performing data analysis.

### Step 1: Import Libraries and Load Data
We will use the **pandas** library to read the CSV file.

```python
import pandas as pd 
df = pd.read_csv("student-mat.csv")
```

### Step 2: Verify Missing Values
Now, we check if there are any blank cells or `NaN` values across the entire dataset.

```python
print("Are there any missing values?", df.isnull().values.any())
```

• Remove duplicates.
### Identify and Count Duplicate Rows
Before removing any records, we check for identical duplicate rows across the entire dataset. We use `df.duplicated().sum()` to find out exactly how many matching rows exist so we don't accidentally drop unique information.

```python
duplicates = df.duplicated().sum()
print("Number of duplicate rows before removal:", duplicates)
```

###  Inspect Dataset Shape & Data Types
Finally, we view the overall dimensions of our grid grid using `df.shape` alongside the structural data types of each column using `df.dtypes`.

```python
print("Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
```

###Analysis Questions
## Average Final Grade (G3)

The average of the final grades (`G3`) was calculated using the `mean()` function in Pandas. This provides an overall measure of students' academic performance in the dataset.

```python
average_grade = df ['G3'].mean()
print("average final grade:",average_grade)
```

## How many students scored above 15?
The count of students scoring above 50 was calculated by count df['G3'] > 15 (The Boolean Mask) .sum() (The Aggregation)

```python
count = (df['G3']>15).sum()
print("no. of students scored above 15:",count)
```

## Is study time correlated with performance?
## Correlation Analysis: Study Time vs. Performance
We use Pearson's correlation coefficient to evaluate the relationship between `studytime` and final grades (`G3`). The result is evaluated using a conditional statement to interpret the direction of the relationship.


```python
correlation = df["studytime"].corr(df["G3"])
print("Is study time correlated with performance?",correlation)
if correlation>0 :
 print("Students who study more tend to score higher (positive correlation)")
elif correlation<0:
 print("Students who study more tend to score lower (negative correlation).")
else: 
 print("There is little or no relationship")
```

## Gender Performance Analysis
Grouping the dataset by the `sex` column to calculate and compare the average final grade (`G3`) between female ('F') and male ('M') students.


```python
gender_performance=df.groupby("sex")["G3"].mean()
print(gender_performance)
F=gender_performance["F"]
M=gender_performance["M"]
if F>M:
 print("Females perform better")
else:
 print(" Males perform better")
```

## Plotting a Grade Histogram
We are using a **histogram** to see how student grades are distributed. This helps us see if most students got high, low, or average marks.

**What the code does:**
* `figsize=(6,4)` sets the size of the graph window.
* `bins=10` divides the grades into 10 groups (bars).
* `color` and `edgecolor` make the bars look clean and easy to read.
* `plt.show()` prints the final graph on the screen.


```python
plt.figure(figsize=(6,4))
plt.hist(df["G3"], bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
## Scatterplot: study time vs grades.
plt.figure(figsize=(6,4))
plt.scatter(df["studytime"], df["G3"], color="green")
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

plt.figure(figsize=(6,4))`**  
    Initializes a new plot canvas with specific dimensions. The `figsize` parameter sets the width to **6 inches** and the height to **4 inches**, ensuring the final graph is well-proportioned.

*   **`plt.scatter(df["studytime"], df["G3"], color="green")`**  
    Plots the data points extracted from the pandas DataFrame (`df`).
    *   `df["studytime"]`: Positioned on the **horizontal X-axis** (independent variable).
    *   `df["G3"]`: Positioned on the **vertical Y-axis** (dependent variable representing the final grade).
    *   `color="green"`: Changes the default blue marker dots to **green**.

*   **`plt.title("Study Time vs Final Grade")`**  
    Places a centralized title at the top of the chart to immediately communicate the objective of the data visualization.

*   **`plt.xlabel("Study Time")` & `plt.ylabel("Final Grade (G3)")`**  
    Applies text labels to the respective axes so viewers can easily identify what the metrics and data ranges represent.

*   **`plt.show()`**  
    Compiles all the previous visual configurations (the canvas, data points, labels, and title) and renders the final graphic on your screen.

```python
## Bar chart: male vs female average score.
gender_avg = df.groupby("sex")["G3"].mean()
plt.figure(figsize=(5,4))
plt.bar(gender_avg.index, gender_avg.values, color=["pink", "lightblue"])
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

   **`gender_avg = df.groupby("sex")["G3"].mean()`**  
    This line handles the data aggregation using pandas before plotting:
    *   `df.groupby("sex")`: Splits the dataset into separate groups based on the unique categories in the "sex" column.
    *   `["G3"]`: Selects the final grade column for calculation.
    *   `.mean()`: Calculates the average final grade for each gender group, saving the result as a pandas Series where the genders are the index and the averages are the values.

*   **`plt.figure(figsize=(5,4))`**  
    Prepares the canvas area. The `figsize` parameter sets the width to **5 inches** and the height to **4 inches**, establishing compact and readable dimensions for a simple categorical comparison.

*   **`plt.bar(gender_avg.index, gender_avg.values, color=["pink", "lightblue"])`**  
    Generates the vertical bar chart:
    *   `gender_avg.index`: Provides the categorical labels for the horizontal X-axis (e.g., 'F' and 'M').
    *   `gender_avg.values`: Sets the heights of the bars corresponding to the calculated average grades.
    *   `color=["pink", "lightblue"]`: Explicitly maps colors to the bars in sequence—the first gender category will be rendered in pink, and the second in light blue.

*   **`plt.title("Average Final Grade by Gender")`**  
    Adds a main title at the top of the chart, clarifying what metric is being compared across the demographic groups.

*   **`plt.xlabel("Gender")` & `plt.ylabel("Average Final Grade (G3)")`**  
    Applies descriptive text labels to the axes to ensure the viewer understands both the categorical groupings (X-axis) and the numerical metric scale (Y-axis).

*   **`plt.show()`**  
    Assembles all configured components (the computed values, colored bars, labels, and title) and cleanly outputs the completed visual chart onto the screen.
